```python
import tensorflow as tf
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import time
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_class_data = pd.read_csv("C:/Users/TANG YOON CHUEN/Desktop/UMS_files/FYP\Dataset/archive/Human Action Recognition/Training_set.csv")
print(train_class_data['label'].unique())
print(len(train_class_data['label'].unique()))
num_unique_classes=len(train_class_data['label'].unique())
```

```python
from keras.preprocessing.image import ImageDataGenerator
idg_train = ImageDataGenerator(rescale=1./255,horizontal_flip=True,shear_range=0.2,zoom_range=0.2,validation_split=0.2)

train = idg_train.flow_from_dataframe(dataframe=train_class_data,
                                              directory="C:/Users/TANG YOON CHUEN/Desktop/UMS_files/FYP/Dataset/archive/Human Action Recognition/train/",x_col='filename',y_col='label',target_size=(64,64),batch_size=64,class_mode='categorical',subset='training')
validation = idg_train.flow_from_dataframe(dataframe=train_class_data,
                                              directory="C:/Users/TANG YOON CHUEN/Desktop/UMS_files/FYP/Dataset/archive/Human Action Recognition/train/",x_col='filename',y_col='label',target_size=(64,64),batch_size=64,class_mode='categorical',subset='validation')
```

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D,Flatten, Dense,BatchNormalization,Dropout
input_shape=(128,128,3)
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
cnn = tf.keras.models.Sequential()

cnn.add(tf.keras.layers.Conv2D(64, kernel_size=3, activation='relu'))
cnn.add(BatchNormalization())
cnn.add(tf.keras.layers.MaxPooling2D(pool_size=2))
cnn.add(Dropout(0.4))

cnn.add(tf.keras.layers.Conv2D(128, kernel_size=3, activation='relu'))
cnn.add(BatchNormalization())
cnn.add(tf.keras.layers.MaxPooling2D(pool_size=2))
cnn.add(Dropout(0.4))

cnn.add(tf.keras.layers.Conv2D(256, kernel_size=3, activation='relu'))
cnn.add(BatchNormalization())
cnn.add(tf.keras.layers.MaxPooling2D(pool_size=2))
cnn.add(Dropout(0.4))

cnn.add(tf.keras.layers.Conv2D(512, kernel_size=3, activation='relu'))
cnn.add(BatchNormalization())
cnn.add(tf.keras.layers.MaxPooling2D(pool_size=2))
cnn.add(Dropout(0.4))

cnn.add(tf.keras.layers.Flatten())

cnn.add(tf.keras.layers.Dense(128, activation='relu'))
cnn.add(BatchNormalization())
cnn.add(Dropout(0.4))

cnn.add(Dropout(0.4))
cnn.add(tf.keras.layers.Dense(num_unique_classes,activation='sigmoid'))


cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
